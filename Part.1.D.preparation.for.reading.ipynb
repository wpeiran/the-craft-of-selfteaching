关于 Python 编程的第一部分总计 7 章，主要内容概括为：

> 1. 了解值的分类和运算方法
> 1. 简要了解流程控制的原理
> 1. 简要了解函数的基本构成

## 阅读策略

首先，不要试图一下子就全部搞懂。这不仅很难，**在最初的时候也完全没必要**。

因为这部分的知识结构中，充满了 “过早引用”。请在第一遍粗略完成第 1 部分中的 E1 ~ E7 之后，再去阅读《[如何从容应对 “过早引用”？](Part.1.F.deal-with-forward-references.ipynb)》。

其次，这一部分，注定要**反复阅读若干遍**。

在开始之前，要明确这一部分的阅读目标。

这一部分的目标，不是让你读完之后就可以开始写程序；也不是让你读完之后就对编程或者 Python 编程有了完整的了解，甚至不是让你真的学会什么…… 这一部分的目标，只是让你 “**脱盲**”。

不要以为脱盲是很容易的事情。你看，所有人出生的时候，都天然是 “文盲”；人们要上好多年的学，才能够真正 “脱盲” —— 仔细想想吧，小学毕业的时候，所有人就真的彻底脱盲了吗？

以中文脱盲为例，学字的同时，还要学习笔划；为了学更多的字，要学拼音，要学如何使用《新华字典》……

学会了一些基础字之后，要学更多的词，而后在练习了那么多造词造句之后，依然会经常用错…… 你看，脱盲，和阅读能力强之间距离很长呢；不仅如此，阅读能力强和写作能力强之间的距离更长……

反复阅读这一部分的结果是：

> * 你对基本概念有了一定的了解
> * 你开始有能力相对轻松地阅读部分官方文档
> * 你可以读懂一些简单的代码

仅此而已。

## 心理建设

当我们开始学习一项新技能的时候，我们的大脑会不由自主地紧张。可这只不过是多年之间在学校里不断受挫的积累效应 —— 学校里别的地方不一定行，可有个地方特别行：给学生制造全方位、无死角、层层递进的挫败感。

可是，你要永远记住两个字：

> 别怕！

用四个字也行：

> 啥也别怕！

六个字也可以：

> 没什么可怕的！

我遇到最多的孱弱之语大抵是这样的：

> 我一个文科生……

哈哈，从某个层面望过去，其实吧，编程既不是文科也不是理科…… 它更像是 “手工课”。你越学就越清楚这个事实，它就好像是你做木工一样，学会使用一个工具，再学会使用另外一个工具，其实总共就没多少工具。然后，你更多做的是各种拼接的工作，至于能做出什么东西，最后完全靠你的想象力……

十来岁的孩子都可以学会的东西，你怕什么？

**别怕**，无论说给自己，还是讲给别人，都是一样的，它可能是人生中最重要的鼓励词。

## 笔记

“别怕”————确实是提高学习质量的最佳策略。

如果在开始学习前不能调整好心态，那么，自己的注意力会不由自主的关注在学习时碰到的问题、麻烦上，“哎呀，真JB麻烦……靠，又那错了？……咋这么难呢？“，并且对它们的恐惧越来越大，大到什么程度呢？大到自己不敢去尝试，不敢去运行某段代码的程度！有这么吓人？

16年自己想尝试学python的时候，使用的是自己的台式机（为了打游戏攒的）装的win7系统（盗版的），然后怎么都安装不了python。然后，通过搜索知道了是系统的问题，且了解到windows对编程不够友好，就想着使用linux类系统。有个情况先说明下，在此之前所有掌握的计算机操作技巧，都是关于如何使用计算机去打游戏的。至于对系统的了解，绝对是小白级别的。

好吧，碰上了个自己之前完全没玩过的情况。第一次按照教程自己用U盘刻了个ubuntu的系统，然后，在开机启动U盘启动时，那个心跳真是……当要安装系统到电脑时，那个犹豫、那个害怕啊————”这要没装好，可就开不机了……win的系统自己还不会装，坏了咋弄？“。就在只能听见自己剧烈的心跳的状态下，第一次安装成了win7+ubuntu双系统……

面对未知自己紧张是难免的，但是绝对有办法去消除、缓解这种恐惧。可以通过问自己这个问题来消解恐惧————**”最坏的情况是什么？“**。就以我给电脑装系统来说，最坏的情况是什么？最坏的情况就是，系统没装好，导致电脑无法开机。大不了花2、30块钱让别人给重装一下不是。你看只要把它想清楚或者知道它是什么样，恐惧的威力就瞬间小了很多很多。

最大的恐惧来自于自己的想象。如果你看过恐怖片的话，你可能注意到一个现象，只要一部电影里那么恐惧的事物在屏幕上一出现、亮相，那么，这部电影对你造成的恐惧感就瞬间减半。只要自己能见到的东西或者已经感受到的疼痛，都让自己有了感知的基础。”哦，就长这样啊？……唉，也没多疼啊……“你反而会对它们有了免疫，可以承受更高的阀值。高级的恐怖片会采取的策略就是”从头到尾都不让你见到那个让自己恐怖的东西“，”想象当中的才是最恐惧的”————不确定就意味着有各种可能不是？你以为的，就真的是你以为的。

到最后，你也会对这个道理“**所有想象中的痛苦都是幻觉**”有深刻的认知和感悟。
